## Disposition Effect
Descripción: Cortar más rapido las pérdidas que las ganancias.

Los inversionistas tienden a vender activos que aumentaron su valor y tienden a mantener los activos cuyo valor ha disminuido.

Analizar las operaciónes a las que pierdes y en las que ganas por sus lapsos de tiempo, revisar cuanto tiempo tomó tenerlas en perdidas y lo mismo con las de las ganancias.

Checar cuando cortó el TP/SL y que tan rápido lo cortaba según iba ganando o perdiendo. Sacar interpretaciones.

Aplicación en caso de trading: De un archivo tradeview se analizan los movimientos realizados, tomando en consideración los trades que fueron cerrados manualmente y los que cerraban al alcanzar el stop loss o take profit.

### Código

#### Se importan librerias necesarias para el funcionamiento correcto del codigo

```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px 
import plotly.graph_objects as go #Libreria necesaria para usar plotly
# import json
```

#### Opciones para visualizar data frames en consola

```python
pd.set_option('display.max_rows', 5000)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

#### Ingreso de datos de entrada: leer archivo json

```python
archivo = '../MyST-Lab3-Equipo4/archivo_tradeview_2.xlsx'
df_datos = pd.read_excel(archivo,
                        header=0,
                        sep=',',
                        parse_dates=False,
                        skip_blank_lines=True)
df_datos.head()
```

```python
df_datos = df_datos.drop(['Order', 'Size', 'Commission', 'Taxes', 'Swap'], 1) #El 1 se refiere a columnas, si fuera 0 serian filas
```

```python
df_datos = df_datos.reset_index()
```

#### Renombramos las columnas en minusculas por convección

```python
df_datos.rename(columns={'Type' : 'type', 'Symbol': 'symbol',
                            'S/L' : 'sl', 'T/P':'tp', 'Profit': 'profit'},
                inplace=True)
df_datos
```

```python
df_times= df_datos[['openTime','closeTime']]
df_times
```

```python
df_times['openTime'] = df_times['openTime'].str.replace('.','-')
df_times['closeTime'] = df_times['closeTime'].str.replace('.','-')
df_times
```

```python
df_datos['openTime'] = df_times['openTime']
df_datos['closeTime'] = df_times['closeTime']
df_datos
```

#### Se realiza una resta entre las fechas que abrio la operación y la cerro

```python
#Ciclo para medir tiempo entre close time y open time manuales de tp y sl

manualTime_tp = []
manualTime_sl = []
for i, row in df_datos.iterrows():
    if row['profit'] > 0 and row['tp'] > 0:
        if row['closePrice']-row['tp'] != 0:
            calc_temp = (datetime.fromisoformat(df_datos.iloc[i,7]).timestamp()
            -datetime.fromisoformat(df_datos.iloc[i,1]).timestamp())/60/60
            manualTime_tp.append(calc_temp)
        else:
            pass
    elif row['profit'] < 0 and row['sl'] > 0:
        if row['closePrice']-row['sl'] != 0:
            calc_temp = (datetime.fromisoformat(df_datos.iloc[i,7]).timestamp()
            -datetime.fromisoformat(df_datos.iloc[i,1]).timestamp())/60/60
            manualTime_sl.append(calc_temp)
        else:
            pass
```

```python
manualTime_tp = pd.DataFrame(manualTime_tp, columns=["manual_tp"])
manualTime_sl = pd.DataFrame(manualTime_sl, columns=["manual_sl"])

```

```python
#Dataframe final
df_manualTimes = pd.concat([manualTime_tp, manualTime_sl], axis=1, sort=False)
df_manualTimes
```

#### Graficar los tiempos que se tardó, con cada uno de los trades.

#### Presentación de grafica: Box Plot

Los extremos de la caja representan los cuartiles inferior y superior, la mediana está marcada por una linea dentro de la caja.


```python
fig = go.Figure()
fig.add_trace(go.Box(
    y=df_manualTimes.iloc[0:,1],
    name='Only Mean',
    marker_color='darkblue',
    boxmean=True # represent mean
))

fig.show()
```
